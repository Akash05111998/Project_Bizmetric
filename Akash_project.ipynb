```python
df_silver=df.join(df1, "driver_id", "inner") \
    .join(df2, "route_id", "inner") \
    .join(df3, "vehicle_id", "inner") \
    .withColumn("fuel_consumed", col("distance_covered") / col("fuel_efficiency"))

display(df_silver)   
# The final DataFrame df_silver contains the joined data from the delivery, driver, route, and vehicle datasets.
# It also includes a new column fuel_consumed, which is calculated by dividing distance_covered by fuel_efficiency. 
```

```python

from pyspark.sql.functions import *

df_silver= df_silver.withColumn("route_name",concat_ws(" to ", col("start_location"), col("end_location")))
df_silver=df_silver.withColumn("processed_date",current_date())

display(df_silver)

# The final DataFrame df_silver contains the joined data from the delivery, driver, route, and vehicle datasets.
# It also includes a new column fuel_consumed, which is calculated by dividing distance_covered by fuel_efficiency.
```
